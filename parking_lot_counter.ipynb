```python
model = YOLO('yolov8n.pt')

results = model.train(
    data='./parking_lot_data/data.yaml',
    epochs=30,
    imgsz=640,
    batch=16,
    patience=10,
    project='runs',
    name='parking_run'
)
```

```python
metrics = model.val()
print("mAP50-95:", metrics.box.map)      # overall mAP, IoU 0.5-0.95
print("mAP50:", metrics.box.map50)       # mAP at IoU 0.5 (easier metric)
print("mAP75:", metrics.box.map75)       # mAP at IoU 0.75 (stricter)
print("Precision:", metrics.box.mp)      # mean precision across classes
print("Recall:", metrics.box.mr)
```

```python
test_metrics = model.val(
    data='./parking_lot_data/data.yaml',
    split='test'
)

print("Test mAP50-95:", test_metrics.box.map)
print("Test mAP50:", test_metrics.box.map50)
print("Test Precision:", test_metrics.box.mp)
print("Test Recall:", test_metrics.box.mr)
```

```python
import gradio as gr
from ultralytics import YOLO
import numpy as np

# Load your trained model (adjust path to your best.pt)
model = YOLO("runs/parking_run-3/weights/best.pt")

def predict(image):
    results = model(image, conf=0.25)
    annotated = results[0].plot()  # returns numpy array (BGR)
    annotated_rgb = annotated[..., ::-1]  # convert to RGB for Gradio

    # Build a summary of detected classes
    names = [model.names[int(c)] for c in results[0].boxes.cls]
    if names:
        from collections import Counter
        counts = Counter(names)
        summary = "\n".join(f"{name}: {count}" for name, count in counts.items())
    else:
        summary = "No objects detected"

    return annotated_rgb, summary

demo = gr.Interface(
    fn=predict,
    inputs=gr.Image(type="numpy", label="Upload an image"),
    outputs=[
        gr.Image(type="numpy", label="Detections"),
        gr.Textbox(label="Class counts")
    ],
    title="YOLO Object Detector",
    description="Upload an image to run detection"
)

demo.launch(share=True, debug=True)
```
